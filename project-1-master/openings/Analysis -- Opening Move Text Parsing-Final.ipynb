## Analysis -- Opening Move Text Parsing

```python
# Import Dependencies 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Parse_Category csv and perform same analysis
parse_test = pd.read_csv("project-1/parse2_categories.csv")
parse_test.head()
```

## Count

```python
opening_category_df = parse_test.groupby(["Opening Category"])
opening_category_count = opening_category_df["Opening Category"].count()
opening_category_count.head()
```

## White Score

Opening Category Results (best openings for White) per Lichess.org games analyzed

```python
opening_category_results = pd.DataFrame(opening_category_df["White Score"].mean())
opening_category_results["Count"] = opening_category_count
filtered_openings = opening_category_results.loc[opening_category_results["Count"] > 1000]
sort_eco = filtered_openings.sort_values("White Score", ascending=False)
sort_eco.iloc[1:10].plot(kind='bar', y="White Score", color="red")
plt.title("Opening Category Results - Best Openings for White", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.savefig("Images/open_cat_white.png")
plt.show()
```

## Black Score

Opening Category Results (best openings for Black as inverse/worst opening for White) per Lichess.org games analyzed

```python
opening_category_results = pd.DataFrame(opening_category_df["White Score"].mean())
opening_category_results["Count"] = opening_category_count
filtered_openings = opening_category_results.loc[opening_category_results["Count"] > 1000]
sort_eco = filtered_openings.sort_values("White Score", ascending=True)
sort_eco.iloc[1:10].plot(kind='bar', y="White Score", color="red")
plt.title("Opening Category Results - Worst Openings for White - Advantage Black", fontsize=9, fontweight="bold")
plt.tight_layout()
plt.savefig("Images/open_cat_black.png")
plt.show()
```

## Build Data Frame to Analyze relationships between Opening Category and Sub-Categories

Opening Sub-Category

```python
opening_category_sub_df = parse_test.groupby(["Opening Sub-Category"])
opening_category_sub_results = pd.DataFrame(opening_category_sub_df["White Score"].mean())
opening_category_sub_count = opening_category_sub_df["Result"].count()
opening_category_sub_results["Count"] = opening_category_sub_count
filtered_openings_sub = opening_category_sub_results.loc[opening_category_sub_results["Count"] > 1000]
sort_sub = filtered_openings_sub.sort_values("White Score", ascending=True)
sort_sub.iloc[1:10].plot(kind='bar', y="White Score", color="red")
plt.show()
```

```python
# Create New data frame with results columns 
opening_category_df = parse_test.groupby(["Opening Category"])
opening_category_results = pd.DataFrame(opening_category_df["White Score"].mean())
opening_category_results
```

Opening

```python
opening_df = parse_test.groupby(["Opening"])
opening_results = pd.DataFrame(opening_df["White Score"].mean())
opening_count = opening_df["Result"].count()
opening_results["Count"] = opening_count
filtered_openings = opening_results.loc[opening_results["Count"] > 1000]
sort_eco = filtered_openings.sort_values("White Score", ascending=True)
sort_eco.iloc[1:10].plot(kind='bar', y="White Score", color="red")
plt.show()
```

```python
# Create New data frame with results columns 
opening_df = parse_test.groupby(["Opening"])
opening_results = pd.DataFrame(opening_df["White Score"].mean())
opening_results
```

```python
# Sort out values 1 and -1
op_df1 = opening_results[opening_results['White Score'] != 1]
op_df2 = op_df1[op_df1['White Score'] != -1]
op_df2
```

```python
# Merge Tables
merge_table = pd.merge(parse_test, op_df2, on='Opening', how='inner')
```

```python
# Merge Tables
merge_table2 = pd.merge(merge_table, opening_category_results, on='Opening Category', how='inner')

# Reorganizing the columns using double brackets
final_df = merge_table2[["Opening","Opening Category","Opening Sub-Category","White Score_y","White Score"]]

# Display
final_df
```

```python
# Rename Columns
renamed_df = final_df.rename(columns={"White Score_y":"Sub-Category Score","White Score":"Category Score"})
renamed_df.head()
```

```python
# Remove Duplicates
renamed_df = renamed_df.drop_duplicates()
renamed_df.head()
```

```python
# Analyze largest differences Opening Category vs. Sub-Category
renamed_df["Difference"] = renamed_df["Sub-Category Score"] - renamed_df["Category Score"]
renamed_df
```
