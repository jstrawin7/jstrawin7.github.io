## Opening Move Text Parsing

DataFrame of openings and their success - analysis of popular variations

Success between variations for a given category of opening

```python
# Import Dependencies 
import pandas as pd
```

```python
# Load a Prexisting csv (from Alex) -- openings and their success
test = pd.read_csv("lichess_2015.csv")
```

```python
# Drop null rows
no_null_test = test.dropna(how='any')
```

```python
# Create New data frame with split value columns 
new = test["Opening"].str.split(':| #|, ', n = 1, expand = True)
```

```python
# Create separate opening category column from New data frame 
test["Opening Category"]= new[0] 
  
# Create separate opening sub-category column from New data frame 
test["Opening Sub-Category"]= new[1] 

# Reorganizing the columns using double brackets
organized_df = test[["Unnamed: 0","White Elo","Black Elo","Result","ECO","White Score","Opening","Opening Category","Opening Sub-Category"]]

# Display
organized_df
```

```python
# Push the New data frame to a new CSV file
organized_df.to_csv("project-1/parse2_categories.csv",
                  encoding="utf-8", index=False, header=True)
```
